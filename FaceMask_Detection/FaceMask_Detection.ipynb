```python
# import libraries

import tensorflow
import matplotlib.pyplot as plt
from matplotlib import pyplot

from tensorflow.keras.preprocessing import image
from tensorflow.keras.layers import Conv2D , MaxPool2D , Flatten , Dense , Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
ls
```

```python
# Define all the paths as variable --

training_path = '/content/drive/MyDrive/Face_Mask_data/Training_set'
testing_path = '/content/drive/MyDrive/Face_Mask_data/Test_set'
prediction_path = '/content/drive/MyDrive/Face_Mask_data/single_prediction'
```

```python
train_datagen = ImageDataGenerator(rescale = 1/255,               #reduce the size
                                   shear_range = 0.2,             #rotate image
                                   zoom_range = 0.2,              # zoom the image
                                   horizontal_flip = True)        #horizontal image
                                   

test_datagen = ImageDataGenerator(rescale = 1/255)               #reduce the size
```

```python
training_set = train_datagen.flow_from_directory(training_path,
                                                 target_size = (64,64),      #dimension
                                                 batch_size = 32,            #to save the memory (RAM) we are taking 32 images at a time then again 32 n so on
                                                 class_mode = 'binary')       #binary classification

test_set = test_datagen.flow_from_directory(testing_path,
                                                 target_size = (64,64),    #dimension (will change image in same size)
                                                 batch_size = 32,
                                                 class_mode = 'binary')   #binary classes we choose bcos we have 2 classes with mask or withput mask


# batch_size = 32 ----->> 32 set of data samples will go for training from datasets at a time.                                                 
```

```python
# 1 define architecture

classifier = Sequential()

# conv layer 1
classifier.add(Conv2D(32 , 3 , activation='relu', input_shape = (64,64,3) ))
classifier.add(MaxPool2D())     #pool

# conv layer 2
classifier.add(Conv2D(32*2 , 3 , activation='relu' ))
classifier.add(MaxPool2D())     #pool

# flatten
classifier.add(Flatten())

# hidden layer 1
classifier.add(Dense(128, activation='relu' , kernel_initializer='he_uniform'))
# classifier.add(Dropout(0.2))

# hidden layer 2
classifier.add(Dense(64, activation='sigmoid'))
# classifier.add(Dropout(0.5))

# hidden layer 3
classifier.add(Dense(32, activation='relu' , kernel_initializer='he_uniform'))
# classifier.add(Dropout(0.5))

#output
classifier.add(Dense(1, activation='sigmoid'))   # we took 1 bcos we want 1 output mask or without mask i.e image is of mask or without mask and we use sigmoid 

#classifier.add(Dense(2, activation='softmax'))   # if more than 1 output we want then we have to use softmax
```

```python
# 2.compile

classifier.compile(optimizer='adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'])
```

```python
# 3. model train

classifier.fit(training_set , epochs = 1 , validation_data = test_set)
```

```python
# acc of training 81 and testing acc 86 ---->> close to each other i.e not overfitted
# both accuracy should be close to each other then we can say that not overfitting
```

```python
# save the model

classifier.save('facemask_classifier_86.h5')          #h5 storing for binary data
```
